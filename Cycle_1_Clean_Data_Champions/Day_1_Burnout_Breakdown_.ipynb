```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore")
sns.set(style="whitegrid")
pd.set_option('display.max_columns', None)
df = pd.read_csv("mental_health_workplace_survey.csv")

df.head()
```

```python
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns

for col in numeric_cols:
    median_val = df[col].median()
    df[col].fillna(median_val, inplace=True)
    print(f"Filled missing values in {col} with median: {median_val}")

categorical_cols = df.select_dtypes(include=['object']).columns

for col in categorical_cols:
    mode_val = df[col].mode()[0]
    df[col].fillna(mode_val, inplace=True)
    print(f"Filled missing values in {col} with mode: {mode_val}")
```

```python
def remove_outliers_iqr(df, col):
    Q1 = df[col].quantile(0.35)
    Q3 = df[col].quantile(0.65)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    original_size = df.shape[1]
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    removed = original_size - df.shape[0]
    print(f"{removed} outliers removed from '{col}'")
    return df

for col in ['TeamSize', 'ProductivityScore', 'WorkHoursPerWeek']:
    df = remove_outliers_iqr(df, col)
```

```python

sns.scatterplot(x='SleepHours', y='StressLevel', hue='BurnoutRisk', data=df, palette='coolwarm')
plt.title("Stress vs Sleep (Colored by BurnoutRisk)")
plt.xlabel("Sleep Hours")
plt.ylabel("Stress Level")
plt.legend(title="BurnoutRisk")
plt.show()
```

```python

sns.countplot(data=df, x='HasMentalHealthSupport', hue='BurnoutRisk', palette='pastel')
plt.title("Mental Health Support vs Burnout Risk")
plt.xlabel("Has Mental Health Support")
plt.ylabel("Count")
plt.legend(title="Burnout Risk")
plt.show()
```
